```python
import networkx as nx
```

# 3.5.1 Depth-First Traversal (DFS)

```python
g = nx.barabasi_albert_graph(10,5)
```

```python
print('Number of nodes', len(g.nodes))
print('Number of edges', len(g.edges))
print('Average degree', sum(dict(g.degree).values()) / len(g.nodes))
```

```python
edges = nx.dfs_edges(g, 0)
```

```python
print(edges)
```

```python
print(list(edges))
```

```python
print(list(nx.dfs_tree(g,0)))
```

```python
tree = nx.dfs_tree(g,0)
print(list(tree.edges()))
```

```python
nx.draw_networkx(tree)
```

```python
successors = nx.dfs_successors(g,0)
print(successors)
```

```python
predecessors = nx.dfs_predecessors(g, 0)
print(predecessors)
```

# 3.5.2 Breadth-First Traversal (BFS)

```python
edges = nx.bfs_edges(g,0)
print(list(edges))
```

```python
print(list(nx.bfs_tree(g,0)))
```

```python
tree = nx.bfs_tree(g, 0)
print(list(tree.edges()))
```

```python
nx.draw_networkx(tree)
```

```python
successors = nx.bfs_successors(g,0)
print(successors)
```

```python
predecessors = nx.bfs_predecessors(g,0)
print(predecessors)
```
